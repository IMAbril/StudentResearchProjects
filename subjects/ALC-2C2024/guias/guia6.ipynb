## Ejercicio 1: Método de Horner

Dado un polinomio $p(x) = a_nx^n+a_{n−1}x^{n−1} + \dots +a_1x+a_0$ .

¿Cuántos productos y cuántas sumas se realizan al evaluar el poliomio en un cierto $x_0$ ?
- $p(x_0) = \sum_{i=0}^n \left( a_i \prod_{j=0}^i x_0 \right) = \sum_{i=0}^n a_ix_0^i \implies$ Tengo n+1 términos en la sumatoria, osea que tengo n sumas. En el i-ésimo término de la sumatoria tengo i multiplicaciones. Luego, tengo $\sum_{i=1}^{n+1}i = \frac{(n+1)(n+2)}{2} = \frac{n^2+3n+2}{2} $ multiplicaciones, le restamos la que cuenta a $x_0^0$como multiplicación y nos quedan $\frac{n^2+3n}{2}$ multiplicaciones.
 
Horner propone como alternativa escribir a $p$ como $p(x)= a_0 + x(a_1 + x(a_2 + \dots + x(a_{n−1} + xa_n )))$.

¿Cuántos productos y cuántas sumas se realizan al evaluar $p$ bajo esta forma?
- Esta forma usa una factorización iterativa del polinomio, desde el coeficiente de mayor grado hasta el menor.
  
  $p(x)= (a_nx^{n-1}+a_{n−2}x^{n−3} + \dots +a_1)x+a_0$

  $p(x)= (a_nx^{n-2}+a_{n−3}x^{n−4} + \dots+a_2)x +a_1)x+a_0$

  $\vdots$

  $p(x)= ((((a_nx +a_{n−1})x+a_{n-2})\cdots)x+a_1)x+a_0$

  Por cada factorización tenemos un producto y una suma, luego como son tantas factorizaciones como coeficientes menos 1 (para $a_n$ ya está factorizado), tenemos n productos y n sumas.
  

Este método justifica que Numpy (y otros lenguajes) utilicen una notación especial para evaluar polinomios, en lugar de usar la notación usual de funciones (f= lambda x: ..., en el caso de Python).

## Ejercicio 2

Sea $A \in \mathbb{R}^{m\times n}$, $m\geq n$, con descomposición DVS reducida $A= \hat{U}\hat{\Sigma}\hat{V}^t$ y pseudo inversa $A^\dagger = \hat{V}\hat{\Sigma}^{-1}\hat{U}^t$. Suponiendo que $rango(A)=n$, probar que $\hat{x}= A^\dagger b$ es la única solución del problema de cuadrados mínimos.


**Demostración:**

- $\hat{x}$ es solución del problema de cuadrados mínimos $\iff z=\hat{x}$ minimiza $||Az-b||_2 \iff A^tA\hat{x}=A^tb$ :

    $\hat{x} = A^\dagger b \implies A^tA\hat{x} = A^tAA^\dagger b = A^t(AA^\dagger)^tb = A^t(A^\dagger)^tA^tb $

    Basta ver que $A^t(A^\dagger)^t = (A^\dagger A)^t = A^\dagger A = I_n$
  
    $A^\dagger A = \hat{V}\hat{\Sigma}^{-1}\hat{U}^t\hat{U}\hat{\Sigma}\hat{V}^t = \hat{V}\hat{\Sigma}^{-1}\hat{\Sigma}\hat{V}^t = \hat{V}\hat{V}^t = I_n $

    $\therefore A^tA\hat{x}=A^tb$
- $\hat{x}$ es solución única:

    Basta ver que $rango(A)=n \implies A^tA $ es invertible.

   $A^tA = (\hat{U}\hat{\Sigma}\hat{V}^t)^t\hat{U}\hat{\Sigma}\hat{V}^t =\hat{V}\hat{\Sigma}^tU^t\hat{U}\hat{\Sigma}\hat{V}^t = \hat{V}\hat{\Sigma}^t\hat{\Sigma}\hat{V}^t = \hat{V}\hat{\Sigma}^2\hat{V}^t$

    Consideramos a $\hat{V}\hat{\Sigma}^2\hat{V}^t$ la DVS de $A^tA$, luego $rango(\hat{\Sigma}^2)=rango(A^tA)=n$ por ser $rango(A)=rango(\Sigma)=n$.

    Como $A^tA \in \mathbb{R}^{n \times n}$ y $rango(A^tA)=n$, se sigue que $A^tA$ es invertible.
  
  $\therefore$el sistema $A^tA\hat{x}=A^tb$ tiene solución única.

    $$
  \begin{pmatrix} 1 && -1 && 1 && -1 \\ 1 && 0 && 0 && 0  \\ 1 && 2 && 4 && 8  \\ 1 && 3 && 9 && 27  \end{pmatrix}\begin{pmatrix}a_0 \\ a_1 \end{pmatrix} = \begin{pmatrix} -1 \\ 3 \\ 11 \\ 27\end{pmatrix}
  $$

  Llamamos $A = \begin{pmatrix} 1 && -1 && 1 && -1 \\ 1 && 0 && 0 && 0  \\ 1 && 2 && 4 && 8  \\ 1 && 3 && 9 && 27  \end{pmatrix}$, $x= \begin{pmatrix} a_0 \\ a_1  \end{pmatrix}$, $b = \begin{pmatrix} -1 \\ 3 \\ 11 \\ 27\end{pmatrix}$. Resta resolver la ecuación normal $A^tAx= A^tb$. Como $rango(A)= 4$ su solución es, por lo visto en el punto anterior,: $x= A^\dagger b$.

```python
# Definición de los datos en NumPy
x1 = np.array([-1, 0, 2, 3])  # Los valores de x
y1 = np.array([-1, 3, 11, 27])  # Los valores de y

# Implementación del método de Horner para evaluar el polinomio
def horner(coefficients, x):
    result = coefficients[-1]
    for coeff in reversed(coefficients[:-1]):
        result = result * x + coeff
    return result

# Datos para graficar el polinomio
x_values = np.linspace(min(x1) - 1, max(x1) + 1, 100)

# Crear una figura con tres subgráficos
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Graficar para grados 1, 2 y 3
for n in range(1, 4):
    # Crear la matriz de Vandermonde con el grado n
    A = np.vander(x1, n + 1, increasing=True)  # La matriz de Vandermonde para grado n

    # Calculo la pseudoinversa usando NumPy
    A_pinv = np.linalg.pinv(A)

    # Obtengo la solución del problema de cuadrados mínimos (coeficientes del polinomio)
    a = A_pinv @ y1

    # Evaluar el polinomio utilizando el método de Horner
    y_values = np.array([horner(a, x_val) for x_val in x_values])

    # Graficar el polinomio ajustado en el subgráfico correspondiente
    axs[n-1].plot(x_values, y_values, label=f'Polinomio de grado {n}', color='blue')
    axs[n-1].scatter(x1, y1, color='red', label='Datos (x, y)')
    axs[n-1].set_xlabel('x')
    axs[n-1].set_ylabel('y')
    axs[n-1].set_title(f'Aproximación polinomial de grado {n} en el sentido de mínimos cuadrados')
    axs[n-1].legend()
    axs[n-1].grid(True)

# Ajustar el espacio entre los subgráficos
plt.tight_layout()

# Mostrar todos los gráficos
plt.show()
```

```python
# Definición de los datos en NumPy
x2 = np.array([-1, 0, 1, 2])  # Los valores de x
y2 = np.array([-3, 1, 1, 3])  # Los valores de y

# Datos para graficar el polinomio
x_values = np.linspace(min(x2) - 1, max(x2) + 1, 100)

# Crear una figura con tres subgráficos
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Graficar para grados 1, 2 y 3
for n in range(1, 4):
    # Crear la matriz de Vandermonde con el grado n
    A = np.vander(x2, n + 1, increasing=True)  # La matriz de Vandermonde para grado n

    # Calculo la pseudoinversa usando NumPy
    A_pinv = np.linalg.pinv(A)

    # Obtengo la solución del problema de cuadrados mínimos (coeficientes del polinomio)
    a = A_pinv @ y2

    # Evaluar el polinomio utilizando el método de Horner
    y_values = np.array([horner(a, x_val) for x_val in x_values])

    # Graficar el polinomio ajustado en el subgráfico correspondiente
    axs[n-1].plot(x_values, y_values, label=f'Polinomio de grado {n}', color='blue')
    axs[n-1].scatter(x2, y2, color='red', label='Datos (x, y)')
    axs[n-1].set_xlabel('x')
    axs[n-1].set_ylabel('y')
    axs[n-1].set_title(f'Aproximación polinomial de grado {n} en el sentido de mínimos cuadrados')
    axs[n-1].legend()
    axs[n-1].grid(True)

# Ajustar el espacio entre los subgráficos
plt.tight_layout()

# Mostrar todos los gráficos
plt.show()
```

## Ejercicio 4

Hallar la constante (polinomio de grado 0) que mejor aproxima en el sentido de cuadrados mı́nimos a una función $f : [a, b] → \mathbb{R}$ en $n$ puntos $x_1,\cdots, x_n$ en $[a, b]$.

como el polinomio es de grado 0, la matriz de vandermonde queda así:

$A = \begin{pmatrix} 1 \\ 1 \\ \vdots \\ 1 \end{pmatrix} \in \mathbb{R}^{n\times 1}$

Luego, para hallar la constante planteo la ecuación normal:

$A^tA a_0 = A^t\begin{pmatrix} f(x_1) \\ f(x_2) \\ \vdots \\ f(x_n) \end{pmatrix} \iff na_0 = \sum_{i=1}^n f(x_i) \iff a_0 = \sum_{i=1}^n \frac{f(x_i)}{n}$

Es decir que el polinomio constante que mejor aproxima en el sentido de cuadrados mínimos a una función es el promedio de los valores esperados.

## Ejercicio 6

Supongamos que se tienen puntos $x_0, x_1,\cdots, x_n$ distintos, y se quiere hallar el polinomio $p$ de grado a lo sumo $n$ de modo que $p(x_i) = y_i$ ($p$ interpola los puntos $(x_i, y_i)$). Plantear la matriz del problema (matriz de Vandermonde) ¿Qué tamaño tiene? Observar que $p$ se puede ver como un caso particular de aproximación en el sentido de cuadrados mı́nimos.

$$
\begin{pmatrix} 1 & x_0 & x_0^2 & \cdots & x_0^n \\ 1 & x_1 & x_1^2 & \cdots & x_1^n \\ \vdots & \vdots & \vdots & \ddots & \vdots \\ 1 & x_n & x_n^2 & \cdots & x_n^n \end{pmatrix} \begin{pmatrix} a_0 \\ a_1 \\ \vdots \\ a_n \end{pmatrix} = \begin{pmatrix} y_0 \\ y_1 \\ \vdots \\ y_n \end{pmatrix}  
$$

La matriz de Vandemonde ($A$) tiene tamaño $(n+1) \times (n+1)$ por tener n+1 datos (tendremos n+1 ecuaciones en el sistema, una por cada dato, cada ecuación representa una de las relaciones que debe cumplir $p$ al ser el polinomio interpolador: $p(x_i)=y_i$, con $p$ de grado a lo sumo $n  \implies p(x_i)= a_0 + a_1x_i + \cdots + a_nx_i^n$ ). Como los $x_i$ son distintos, la matriz es invertible, luego el polinomio interpolador es solución del problema de cuadrados mínimos si se busca aproximar con un polinomio del mismo grado que el polinomio interpolador, ya que $ \exists A^{-1} \implies \text{el sistema tiene solución y es única }: a = A^{-1}y \implies ||Aa-y|| = 0 $ la solución del problema del polinomio interpolador también es solución del problema de mínimos cuadrados.

*En síntesis*: El polinomio interpolador puede verse como un caso particular de mínimos cuadrados en el que el error es cero ($||Aa−y||=0$). Esto ocurre cuando la matriz $A$ de Vandermonde en el problema de cuadrados mínimos es cuadrada e inversible, es decir cuando se tienen tantos datos como el grado, a lo sumo, del polinomio con el que se busca aproximar menos 1 (datos: $n+1$ y grado a lo sumo:$n$ garantiza que $A$ sea cuadrada) y todos los datos $x_i$ son únicos (garantiza que $A$ sea inversible). El problema coincide con el de polinomio interpolador.

## Ejercicio 7

Interpolar la siguiente función en $n+1$ puntos equiespaciados en el intervalo $[-1,1]$.  
$$
f(x)= \frac{1}{1+25x^2}
$$

Graficar simultáneamente la función con sus respectivos interpoladores $p_n$ para $n=5, 10, 15$. Para valores altos de $n$ ¿mejora o empeora la aproximación? Calcular la norma infinito de $(f-p_n)$

```python
def f(x_list):
    y_list = []
    for x in x_list:
        y = 1/(1+25*(x**2))
        y_list.append(y)
    return y_list

n_array = [5, 10, 15]
x_array = []
y_array = []
coeficientes = []

for n in n_array:
    x = np.array(np.linspace(-1, 1, n+1)) #obtengo n+1 valores de x
    y = np.array(f(x)) #obtengo n+1 valores de y
    x_array.append(x)
    y_array.append(y)
    A = np.vander(x, n+1, increasing=True)  # La matriz de Vandermonde para grado n con los datos x
    A_inv = np.linalg.inv(A)
    a = A_inv @ y.T #Resuelvo el sistema Aa=y
    coeficientes.append(a) #Obtengo los coeficientes del polinomio interpolador para cada n

# Rango de x para graficar la función y los polinomios
x_dense = np.linspace(-1, 1, 500)  # Mucho más denso para graficar

# Graficar f(x)
plt.figure(figsize=(10, 6))
plt.plot(x_dense, f(x_dense), label='$f(x) = 1/(1+25x^2)$', color='black', linewidth=2)

# Graficar los polinomios interpoladores
colors = ['red', 'blue', 'green']
for i, n in enumerate(n_array):
    a = coeficientes[i][::-1]  # Invertir para usar con np.polyval
    P_n = np.polyval(a, x_dense)  # Evaluar el polinomio en los puntos densos
    plt.plot(x_dense, P_n, label=f'$P_{{{n}}}(x)$', color=colors[i])

# Personalizar la gráfica
plt.title('Función original y polinomios interpoladores', fontsize=14)
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$y$', fontsize=12)
plt.axhline(0, color='gray', linewidth=0.5, linestyle='--')
plt.axvline(0, color='gray', linewidth=0.5, linestyle='--')
plt.legend(fontsize=10)
plt.grid(alpha=0.3)
plt.show()
    
```

Para valores altos de $n$ la aproximación empeora ya que comienza a oscilar y el error se hace más grande.

$||f−p_n||_\inf​= \max_{x\in[-1,1]}​||f(x)−p_n(x)|∣$

```python
#Evaluo las funciones y calculo las diferencias 
f_values = f(x_dense)
p_5values = np.polyval(coeficientes[0][::-1], x_dense)
p_10values = np.polyval(coeficientes[1][::-1], x_dense)
p_15values = np.polyval(coeficientes[2][::-1], x_dense)

differences_p5 = np.abs(f_values - p_5values)
differences_p10 = np.abs(f_values - p_10values)
differences_p15 = np.abs(f_values - p_15values)

# Calcular la norma infinito
norm_infinityp5 = np.max(differences_p5)
norm_infinityp10 = np.max(differences_p10)
norm_infinityp15 = np.max(differences_p15)

# Mostrar resultados en formato tabular
print("n    norma infinito")
print("------------------------------------")
for n, norm in zip(n_array, [norm_infinityp5, norm_infinityp10, norm_infinityp15]):
    print(f"{n:<6} {norm:.2f}")
```

Este polinomio resuelve el problema de mínimos cuadrados porque su grado es la cantidad de datos menos 1 (5)y los datos $x$ son todos distintos.

```python
# Datos
x = np.array([1950, 1960, 1970, 1980, 1990, 2000])  # Año
g_vals = np.array([17, 20.5, 23.9, 27.9, 32.6, 36.9])   # Hab. (millones)
p_vals =  np.array([np.log(17), np.log(20.5), np.log(23.9), np.log(27.9), np.log(32.6), np.log(36.9)])

coeffs_deg5 = np.polyfit(x, p_vals, 5)  # Coeficientes de p(x) grado 5
p_deg5 = np.poly1d(coeffs_deg5)  # Polinomio p(x) grado 5

# Crear la variable simbólica
x_sym = symbols('x')

# Convertir los coeficientes en un polinomio simbólico
polinomio_simbolico = Poly(coeffs_deg5, x_sym) 

# Mostrar el polinomio en formato bonito
print("El polinomio de grado 5 es:")
display(polinomio_simbolico)

coeffs_deg1 = np.polyfit(x, p_vals, 1)  # Coeficientes de p(x) grado 1
p_deg1 = np.poly1d(coeffs_deg1)  # Polinomio p(x) grado 5

# Convertir los coeficientes en un polinomio simbólico
polinomio_simbolico = Poly(coeffs_deg1, x_sym) 

# Mostrar el polinomio en formato bonito
print("El polinomio de grado 1 es:")
display(polinomio_simbolico)
```

```python
# Reconstruir las funciones g(x) = e^(p(x))

g_deg5 = lambda x: np.exp(p_deg5(x))
g_deg1 = lambda x: np.exp(p_deg1(x))

# Puntos para graficar
x_dense = np.linspace(1950, 2000, 500)  

# Evaluación de las funciones ajustadas
g_deg5_values = g_deg5(x_dense)
g_deg1_values = g_deg1(x_dense)

# Gráficos
plt.figure(figsize=(12, 6))

# Datos originales
plt.scatter(x, g_vals, color='black', label='Datos originales', zorder=5)

# Ajuste grado 5
plt.plot(x_dense, g_deg5_values, label='$g(x) = e^{p_5(x)}$', color='blue', linestyle='--')

# Ajuste grado 1
plt.plot(x_dense, g_deg1_values, label='$g(x) = e^{p_1(x)}$', color='red', linestyle='--')

# Personalización del gráfico
plt.title('Ajuste exponencial con polinomios de grado 5 y 1', fontsize=14)
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$g(x)$', fontsize=12)
plt.legend(fontsize=10)
plt.grid(alpha=0.3)
plt.show()
```

Aunque es un polinomio de grado 5, muchos de los coeficientes de los términos superiores ($x^5,x^4,x^3$) son muy pequeños, lo que sugiere que la contribución principal viene de los términos de menor grado. Por lo tanto el comportamiento del ajuste es similar a un polinomio de grado más bajo. Esto se comprueba comparando las gráficas de $g$ aproximada con $p_1$ y $g$ aproximada con $p_5$.

## Ejercicio 11

Implementar un programa que reciba como input una lista de funciones $\{f_1, \cdots, f_m \}$ y dos vectores $x = (x_1 , \cdots, x_n), y = (y_1 , \cdots, y_n)$, y calcule la función $f(x)= \sum_{i=1}^n \alpha_i f_i(x) $ que mejor aproxime los datos en el sentido de cuadrados mı́nimos.

$\implies$ buscamos los coeficientes $\alpha_i$ que minimicen $||f-y||^2$.

Queremos expresar a $f$ como el producto entre una matriz donde cada entrada depende de una evaluación específica y el vector de coeficientes.

$(f(x))_j = \sum_{i=1}^n \alpha_i (f_i(x))_j$

$\implies f = A \begin{pmatrix}\alpha_1 \\ \alpha_2 \\ \vdots \\ \alpha_m \end{pmatrix} = \begin{pmatrix} (f_1)_1 & (f_2)_1 & \cdots & (f_n)_1 \\ (f_1)_2 & (f_2)_2 & \cdots & (f_n)_2 \\ \vdots & \vdots & \ddots & \vdots \\ (f_1)_n & (f_2)_n & \cdots & (f_n)_n \end{pmatrix} \begin{pmatrix}\alpha_1 \\ \alpha_2 \\ \vdots \\ \alpha_n \end{pmatrix}$

Resta resolver las ecuaciones normales, 

$A(x)^tA(x) \alpha = A(x)^ty$


```python
def aproximar_datos_con_combinacionLinealFunciones(x, y, funciones):
    #Construyo la matriz de funciones evaluadas
    A = np.column_stack([f(x) for f in funciones])
    #Resuelvo la ecuación normal
    alpha = np.linalg.solve(A.T @ A, A.T @ y) 
    #Resconstruyo la funcion de ajuste con los coeficientes encontrados
    fitted_function = lambda x_new: sum(alpha[i] * funciones[i](x_new) for i in range(len(funciones)))
    return alpha, fitted_function
```

La ecuación general correspondiente a un modelo de regresión lineal es: $\beta_0 + \sum \beta_kX_k + \epsilon$, donde $\beta$ representa las estimaciones de parámetros lineales que se deben calcular y $\epsilon$ representa los términos de error.

En este caso planteamos un modelo de regresión lineal simple porque tenemos un único predictor (edad gestacional).
