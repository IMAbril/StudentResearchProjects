```python
A = Matrix([[1, -2, 2], [-1,2,-2]])
lamda = symbols('lamda')

#Construcción de V
M = A.T * A
p = M.charpoly(lamda)
factor(p.as_expr())

avecs = M.eigenvects()
avecs = [avecs[0][2][0], avecs[0][2][1], avecs[1][2][0]]

bon = GramSchmidt(avecs, orthonormal=True)
V = bon[0].row_join(bon[1]).row_join(bon[2])
V
```
